```python
#Construindo array de allow e disallow do robots.txt
###Duvida: Levar em consideração Visit-time e Crawl-delay do robots ?
def criarDisallowArray(pageCounter):
    robotsTXT = 'robots/' + sites[pageCounter][0] + ".txt"
    with open(robotsTXT) as f:
       content = f.readlines()

    start = len(content)
    end = len(content)
    disallowArray = []
    allowArray = []
    visitTime = ''
    crawlDelay = ''
    for i in range(len(content)):
       if('User-agent: *' in content[i]):
           start = i
           break
    for i in range(start+1, len(content)):
       if('User-agent: ' in content[i]):
           end = i
           break
    for i in range(start+1, len(content)):
       if('Visit-time: ' in content[i]):
           visitTime = content[i].split(' ')[1]
           break
    for i in range(start+1, len(content)):
       if('Crawl-delay: ' in content[i]):
           crawlDelay = content[i].split(' ')[1].split('\n')[0]
           break
    for i in range (start+1, end-1):
       line = content[i].split(' ')
       if(line[0] == 'Allow:'):
           allowArray.append(line[1].split('\n')[0])
       elif(line[0] == 'Disallow:'):
           disallowArray.append(line[1].split('\n')[0])
    return disallowArray
```

1 - Abordagem BFS

```python
#Funções auxiliares 
def isAllowed(disallowArr,url):
    for i in range (len(disallowArr)):
        if(disallowArr[i] in url):
            return False
    return True
```

```python
def isRootSite(siteUrl):
    return not('http://' in siteUrl or 'https://' in siteUrl)
```

```python
def getLinks(page):
    parsedPage = BeautifulSoup(page,"html.parser")
    anchors = parsedPage.find_all('a', href = True)
    links = []
    for anchor in anchors:
        links.append(anchor['href'])
    return links
```

```python
def getPage(urlPage):
    req = rq.get(urlPage)
    return req.text
```

```python
def downloadPage(urlPage,siteFolder,pageCounter,writePage):
    try:
        req = rq.get(urlPage)
        if(writePage==1):
            directoryFile = 'pagesHeuristica/' + siteFolder + "/page" + str(pageCounter) + ".html" 
            os.makedirs(os.path.dirname(directoryFile), exist_ok=True)
            with open(directoryFile, 'wb') as handle:
                for block in req.iter_content():
                    handle.write(block)
        return req.text
    except rq.exceptions.Timeout:
        print("Timeout occurred")
        
```

2 - Abordagem com heuristicas

```python
#A heuristica usada aqui será baseada em evitar ir para outros esportes e buscar 
#sempre urls que contenham palavras-chave para chegar em páginas de jogadores

# sites = [('MLSSoccer','https://www.mlssoccer.com'),
# ('Eurosport','https://www.eurosport.com'),
# ('Soccerway','https://us.soccerway.com'),
# ('FCTable','https://www.fctables.com'),
# ('FIFA','https://www.fifa.com'),
# ('Fbref','https://fbref.com'),
# ('UEFA','https://www.uefa.com')]

#Termos que não são desejados => nomes de outros esportes, informações genericas (news,history,about,...)
#Termos que normalmente indicam proximidade => roster, squad, team(s),club(s), rotas terminadas em nomes de times
#Termos que normalmente indicam paginas desejadas => rotas com player(s),person
notWantedSports = ['nfl','nba','nhl','mlb','baseball','tennis','athletics','basketball',
                   'boxing','cricket','cycling','swimming','olympics','equestrian',
                   'judo','volleyball','rugby','formula-1','golf','ice-hockey','bobsleigh',
                   'curling','wrc','moto','formula-e','wtcr']
notWantedTerms = ['destination','organisation','match','ticket','development','awards',
                  'stories','internal','fan','cookie','archive','video','photo','statistics',
                  'venue','trophies','post','news','about','history','rules','ios','android',
                  'twitter','instagram','facebook','customer','media','jobs','work','coach',
                  'story','live','calendar','executive','series','meta','results','info','stat',
                  'comments','profile','trophy','legend','library','browser']
wantedTerms = ['/player','/person']
closeTerms = ['cup','league','roster','squad','team','tournaments']
highlyWantedUrls = ['/players/player/','/player=','/players/','/person']
def wantedLink(link):
    for nws in notWantedSports:
        if(nws in link):
            return -1
    for nwt in notWantedTerms:
        if(nwt in link):
            return -1
    for wt in wantedTerms:
        if(wt in link):
            return 2
    for ct in closeTerms:
        if(ct in link):
            return 1
    return 0

def removeTrash(lista):
    return filter(lambda x : x[1]>0 , lista)
```

```python
import queue
#for i in range (len(sites)):
i=6
urlQueue = queue.Queue()
visited = set()
downloaded = []
disallowArrayToUse = criarDisallowArray(i)
siteFolder = sites[i][0]
siteRootUrl = sites[i][1]
visited.add(siteRootUrl)
urlQueue.put(siteRootUrl)
pageCounter = 0;
adicionado = [];
descartado = [];
while((not urlQueue.empty()) and pageCounter<10): 
    currentUrl = urlQueue.get()
    #print(currentUrl)
    page = downloadPage(currentUrl,siteFolder,pageCounter)
    downloaded.append(currentUrl)
    pageCounter+=1
    #print(pageCounter)
    for url in getLinks(page):
        if ((siteRootUrl+url not in visited) and isRootSite(url) and isAllowed(disallowArrayToUse,url) and (url.startswith('/')) and ('{' not in url) and (':' not in url)):
            wl = wantedLink(url)
            if(wl>0):
                visited.add(siteRootUrl+url)
                urlQueue.put(siteRootUrl+url)
                #print("Adicionado "+url)
                adicionado.append([url,wl])
            else:
                descartado.append([url,wl])
                #print("Descartado "+url)
#downloaded
adicionado
#descartado
```
